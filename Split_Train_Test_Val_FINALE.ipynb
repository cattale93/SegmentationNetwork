```
#set up config param
prc_train = 0.7
prc_test = 0.2
prc_valid = 0.1


#set up target folder

processed_data = '/content/drive/My Drive/Seg_net/dataset/lists/'
splitted_data = '/content/drive/My Drive/Seg_net/dataset/down/'
os.chdir(processed_data)
folder = ['2_Patch_size64_prc10_25_overlap_Thu_Jan_30_17_17_12_2020']

for cartella in folder:

    os.chdir(processed_data + cartella)

    data = np.load('data/data.npy')
    tare = np.load('tare/tare.npy')

    n_patches = data.shape[0]
    indexes = np.array(range(0,n_patches))
    np.random.shuffle(indexes)



    for batch_size in [32]:

        print(cartella, ': ', batch_size)

        n_train = int (n_patches * prc_train)
        n_train = n_train + (batch_size - n_train % batch_size)
        n_test = int(n_patches * prc_test)
        n_test = n_test + (batch_size - n_test % batch_size)
        n_valid = n_patches - n_train - n_test

        print('n_patches {}, n_train {}, n_test {}, n_valid {}'.format(n_patches,n_train,n_test,n_valid))
        print('Avanzo train {}, test {}, valid {}'.format(n_train % batch_size, n_test % batch_size, n_valid % batch_size))
        

        try:

            os.chdir(splitted_data + cartella)
            
            temp = data[indexes[0:n_train],:,:,:]
            np.save('train' + str(batch_size) + '/data/data.npy', temp)
            
            temp = data[indexes[n_train:n_train + n_test],:,:,:]
            np.save('test' + str(batch_size) + '/data/data.npy', temp)

            temp = data[indexes[n_train + n_test:],:,:,:]
            np.save('valid' + str(batch_size) + '/data/data.npy', temp)

            temp = tare[indexes[0:n_train],:,:,:]
            np.save('train' + str(batch_size) + '/tare/tare.npy', temp)

            temp = tare[indexes[n_train:n_train + n_test],:,:,:]
            np.save('test' + str(batch_size) + '/tare/tare.npy', temp)

            temp = tare[indexes[n_train + n_test:],:,:,:]
            np.save('valid' + str(batch_size) + '/tare/tare.npy', temp)


        except:
            print('dafdas')
```
